```python
stock_data1 = pd.read_csv('STOCK1.csv')
stock_data1
```

```python
stock_data2 = pd.read_csv('STOCK2.csv')
stock_data2
```

```python
pd.concat([stock_data1,stock_data2],axis=0)##axis=0 為直向合併
```

```python
pd.concat([stock_data1,stock_data2],axis=1)
```

```python
pd.concat([stock_data1,stock_data2],axis=0,join='inner')##inner是交集
```

```python
pd.concat([stock_data1,stock_data2],axis=0,join='outer')#outer是聯集
```

```python
pd.merge(stock_data1,stock_data2,how='outer')
```

```python
pd.merge(stock_data1,stock_data2,on='date',how='outer')
```

```python
pd.concat([stock_data1,stock_data2],axis=1)##axis=1 為橫向合併
```

```python
pd.merge(stock_data1,stock_data2,on='date',how='left')
```

```python
pd.merge(stock_data1,stock_data2,on='date',how='right')
```

```python
stock_data1_index = stock_data1.set_index('date')
stock_data2_index = stock_data1.set_index('date')
```

```python
stock_data1_index.join(stock_data2_index,how='outer',lsuffix='_left',rsuffix='_right')
```
